```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn import linear_model

df = pd.read_csv("honeyproduction.csv")
```

```python
prod_per_year=df.groupby('year').totalprod.mean().reset_index()
print(prod_per_year)
```

```python
X=prod_per_year['year']
X=X.values.reshape(-1,1)

y=prod_per_year['totalprod']
y=y.values.reshape(-1,1)
```

```python
plt.scatter(X,y)
plt.show()
```

```python
line_fitter=linear_model.LinearRegression()
line_fitter.fit(X,y)
line_fitter.coef_,line_fitter.intercept_
y_predict=line_fitter.predict(X)
plt.xlabel("Year")
plt.ylabel("Total honey production")
plt.title("Honey production per year")
plt.scatter(X,y)
plt.plot(X,y_predict)
plt.show()
```

```python
x_future=np.array(range(2013,2051))
x_future=x_future.reshape(-1,1)
future_predict=line_fitter.predict(x_future)
plt.xlabel("Year")
plt.ylabel("Total honey production")
plt.title("Honey production per year")
plt.scatter(X,y)
plt.plot(X,y_predict)
plt.plot(x_future,future_predict)
plt.show()
```
